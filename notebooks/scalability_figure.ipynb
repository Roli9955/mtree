```python
BASE_FIGURE_SIZE = 3.0
GOLDEN_RATIO = 1.618033988749894
MARKER_SIZE = 40
```

```python
from matplotlib import pyplot as plt
import os

row_numbers = [rn for rn in range(10000, 100001, 10000)]

mtree_float_y = [213,254,272,340,395,512,553,656,754,788]
mtree_int8_y  = [202,227,285,372,517,616,751,875,1055,1109]
mtree_text_y  = [1244,2517,4021,5225,6527,7775,9184,10735,11993,13444]

plt.figure(figsize=(BASE_FIGURE_SIZE * GOLDEN_RATIO, BASE_FIGURE_SIZE), layout='constrained')
plt.grid(False)

plt.scatter(row_numbers, mtree_float_y, MARKER_SIZE, marker='x')
plt.scatter(row_numbers, mtree_int8_y,  MARKER_SIZE, marker='o')
plt.scatter(row_numbers, mtree_text_y,  MARKER_SIZE, marker='v')

plt.plot(row_numbers, mtree_float_y, label='mtree_float')
plt.plot(row_numbers, mtree_int8_y,  label='mtree_int8')
plt.plot(row_numbers, mtree_text_y,  label='mtree_text')

plt.xlabel('Indexelt sorok száma')
plt.ylabel('Végrehajtási idő (milliszekundum)')
plt.legend()

plt.savefig(os.path.expanduser('~/Downloads/mtree_create_index.pdf'))
```
