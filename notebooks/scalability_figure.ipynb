```python
from matplotlib import pyplot as plt
import numpy as np
import os

tags     = ['A nyers adat\nmérete','A tábla mérete\naz indexelés előtt','Az index mérete', 'A tábla mérete\naz indexelés után']
bytes    = [292686977,521797632,601997312,1123958784]
bytes_mb = [((b / 1024) / 1024) for b in bytes]

y_position = np.arange(len(tags))

plt.figure(figsize=(BASE_FIGURE_SIZE * GOLDEN_RATIO * GOLDEN_RATIO, BASE_FIGURE_SIZE), layout='constrained')
plt.grid(False)

plt.bar_label(plt.barh(y_position, bytes_mb, align='center'), labels=['%.2f' % mb for mb in bytes_mb], padding=8, fontsize=12, label_type='center', color='w')
plt.yticks(y_position, labels=tags)

plt.xlabel('Lemezen elfoglalt tárhely (megabájt)')
plt.legend()

plt.savefig(os.path.expanduser('~/Downloads/mtree_storage_big_data.pdf'))
```
